```python
class Person:
  def __init__(self,fname,lname,age):
    self.fname = fname
    self.lname = lname
    self.age = age
  def display(self):
    print(f'fname: {self.fname}, lname: {self.lname}, age: {self.age}')
  def update(self,fname,lname,age):
    self.fname = fname
    self.lname = lname
    self.age = age

class Employee(Person):
  def __init__(self,fname,lname,age,post):
    super().__init__(fname,lname,age)
    self.post = post
  def update_emp(self,fname,lname,age,post):
    self.update(fname,lname,age)
    self.post = post
  def display_emp(self):
    self.display()
    print(f'post: {self.post}')

x = Employee('a','b','32','rab')
x.display()
```

```python
class Film:
  def __init__(self,name,rating):
    self.name = name
    self.rating = rating

class Cinema(Film):
  def __init__(self,adress):
    self.adress = adress
    self.films = []
  def add_film(self,name,rating):
    new_film = Film(name,rating)
    self.films.append(new_film)
  def display(self):
    print(f'adres: {self.adress}')
    print('Films:')
    for i in self.films:
      print(f'Film: {i.name}, rating: {i.rating}')

x = Cinema('Baumana 4')
x.add_film('Barbi',8)
x.add_film('Draiv',9)
x.display()
```

```python
class Project:
  def __init__(self,project,deadline):
    self.project = project
    self.deadline = deadline

class Employee(Project):
  def __init__(self,name):
    self.name = name
    self.projects = []
  def add_project(self,project):
    if project not in self.projects:
      self.projects.append(project)
  def display(self):
    print(f'Employee: {self.name}')
    print('Projects:')
    for i in self.projects:
      print(f'project name: {i.project}, deadline: {i.deadline}')

x = Employee('Oguzok')
p1 = Project('morda', '29 february')
p2 = Project('lada kalina', '2025 year')
x.add_project(p1)
x.add_project(p2)
x.display()
```

```python
from abc import ABC, abstractmethod

class ProjectTask(ABC):
  def __init__(self,name):
    self.name = name

  @abstractmethod
  def startTask(self):
    pass
  @abstractmethod

  def completeTask(self):
    pass

class DevolopmentTask(ProjectTask):
  def startTask(self):
     print(f'Start of devolopment {self.name}')
  def completeTask(self):
     print(f'Devolopment of {self.name} completed')

class TestingTask(ProjectTask):
  def startTask(self):
     print(f'Testing of {self.name}')
  def completeTask(self):
     print(f'Testing of {self.name} completed')

dtask = DevolopmentTask('уронить цру')
ttask = TestingTask('проверить упал ли')

dtask.startTask()
dtask.completeTask()

ttask.startTask()
ttask.completeTask()
```

```python
from abc import ABC, abstractmethod

class Employee(ABC):
  def __init__(self, name, salary):
    self.name = name
    self.salary = salary
  @abstractmethod
  def calculateBonus(self):
    pass
  def display(self):
    print(f'Name: {self.name}, Salary: {self.salary}, Bonus: {self.calculateBonus()}')

class FullTimeEmployee(Employee):
  def calculateBonus(self):
     return self.salary * 2

class PartTimeEmployee(Employee):
  def calculateBonus(self):
     return self.salary * 1.5

x = FullTimeEmployee('aleg', 75000)
y = PartTimeEmployee('zhenya', 42000)
x.display()
y.display()
```
