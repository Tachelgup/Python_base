```python
def word_frequency(text):
    words = text.lower().split()
    freq_dict = {}
    for word in words:
        freq_dict[word] = freq_dict.get(word, 0) + 1
    return freq_dict

text = input('Enter text: ')
result = word_frequency(text)
print(result)
```

```python
from random import randint

def new_dict(freq_dict, pivot):
  new_d = {}
  for key, value in freq_dict.items():
    if value > pivot:
      new_d[key] = value
  return new_d

n = randint(4,10)
freq_dict = {}
for i in range(n):
  freq_dict[i] = randint(-100, 100)

print("Dir:", freq_dict)
pivot = int(input('Pivot: '))
result = new_dict(freq_dict, pivot)
print("New dir:", result)
```

```python
from random import randint

def pop_keys(dict, keys):
  new_dict = dict.copy()
  for key in keys:
    new_dict.pop(key, None)
  return new_dict

n = randint(4,10)
freq_dict = {}
for i in range(n):
  freq_dict[i] = randint(-100, 100)
keys = [1, 3]

print("Dir:", freq_dict)
print("New dir: ", pop_keys(freq_dict,keys))
```

```python
from random import randint

def merge(dict1, dict2):
  new_dict = {}
  for key, value in dict1.items():
    new_dict[key] = [value]
  for key, value in dict2.items():
    if key in new_dict:
      new_dict[key].append(value)
    else:
      new_dict[key] = [value]
  return new_dict

n = randint(4,10)
dict1 = {}
dict2 = {}
for i in range(n):
  dict1[i] = randint(-5, 5)
m = randint(4,10)
for i in range(m):
  dict2[i] = randint(-5, 5)

print("Dir1:", dict1)
print("Dir2:", dict2)
print("Merged:", merge(dict1, dict2))
```

```python
from random import randint

def mult_val(dict):
  mult = 1
  for key in dict:
    mult *= dict[key]
  return mult

n = randint(2,5)
freq_dict = {}
for i in range(n):
  freq_dict[i] = randint(1, 10)

print("Dir:", freq_dict)
print("Mult values:", mult_val(freq_dict))
```

```python
from random import randint

def unique_val(dict):
  uniq = set()
  result = {}
  for key, value in dict.items():
    if value not in uniq:
      result[key] = value
      uniq.add(value)
  return result

n = randint(5,10)
freq_dict = {}
for i in range(n):
  freq_dict[i] = randint(1, 10)

print("Dir:", freq_dict)
print("Unique values:", unique_val(freq_dict))
```

```python
from random import randint

def add_val(dict, val):
  for key in dict:
    dict[key] += val
  return dict

n = randint(5,10)
freq_dict = {}
for i in range(n):
  freq_dict[i] = randint(1, 10)

print("Dir:", freq_dict)
print(f"Values + {n}:", add_val(freq_dict, n))
```

```python
from random import randint

def count_val(dict):
  result = {}
  for key, value in dict.items():
    if value in result:
      result[value] += 1
    else:
      result[value] = 1
  return result

n = randint(5,10)
freq_dict = {}
for i in range(n):
  freq_dict[i] = randint(1, 5)

print("Dir:", freq_dict)
print(f"Count of values:", count_val(freq_dict))
```

```python
from random import randint

def uniq(list1,list2):
  set1 = set(list1)
  set2 = set(list2)
  return set1 - set2

n = randint(5,10)
list1 = []
list2 = []
for i in range(n):
  list1.append(randint(1, 5))
  list2.append(randint(1, 5))

print("List1:", list1)
print("List2:", list2)
print("uniq:", uniq(list1,list2))
```

```python
from random import randint

def rate(list):
  uniq = set(list)
  freq = {}
  for elem in uniq:
    freq[elem] = list.count(elem)
  return freq

n = randint(5,10)
list = []
for i in range(n):
  list.append(randint(1, 5))

print("List:", list)
print("res:", rate(list))
```

```python
from random import randint

def even_odd(set1):
  even = set()
  odd = set()
  for elem in set1:
    if elem % 2 == 0:
      even.add(elem)
    else:
      odd.add(elem)
  return even, odd

n = randint(5,10)
set1 = set()
for i in range(n):
  set1.add(randint(1, 5))

print("Set1:", set1)
even, odd =  even_odd(set1)
print("even:",even)
print("odd:",odd)
```

```python
from random import randint, choice

def rand_del(set1):
  element = choice(list(set1))
  set1.remove(element)
  return set1

n = randint(5,10)
set1 = set()
for i in range(n):
  set1.add(randint(1, 5))

print("Set1:", set1)
print("pop rand:",rand_del(set1))
```

```python
from random import randint

def even_odd_len(set1):
  even = set()
  odd = set()
  for elem in set1:
    if len(elem) % 2 == 0:
      even.add(elem)
    else:
      odd.add(elem)
  return even, odd

list = ['a', 'ab', 'abc', 'abcd', 'abcde']
set1 = set(list)

print("Set1:", set1)
even, odd =  even_odd_len(set1)
print("even:",even)
print("odd:",odd)
```
