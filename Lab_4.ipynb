```python
str = input('string: ')
s = 0
number = ''
flag = False
for i in str:
  if i.isdigit():
    number += i
    flag = True
  elif flag:
    s += int(number)
    flag = False
    number = ''
if flag: s += int(number)
print(f'summ = {s}')
```

```python
str = input('string: ')
result = ''
for i in str:
  if i == ' ': result += ', '
  else: result += i
print(result)
```

```python
str = input('string: ')
count = 0
str = str.split('.')[0]
words = str.split(',')
for i in words:
  if i[0] == i[-1]: count += 1
print(f'count = {count}')
```

```python
FILENAME = "/content/file2.txt"
numbers = []
with open(FILENAME, "r") as file:
    for line in file:
        numbers.extend(map(int, line.split()))
N = len(numbers)
res = []

for i in range((N + 1) // 2):
    res.append(numbers[i])
    if N - 1 - i > i:
        res.append(numbers[N - 1 - i])
print(res)
```
